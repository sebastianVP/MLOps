# **MLOps:**
- Es un conjunto de practicas no es una biblioteca o una herramienta, es un conjunto de practicas que tienen como objetivo implementar y  mantener los modelos de aprendizaje  automatico en produccion de manera confiable y eficiente para asegurararse de que si hay algun cambio en los datos que vuelve a entrar el modelo hace confiable la produccion. ## **INTRODUCTION TO PIPELINES & STEPS**
- ZenML tiene un enfoque basado en una tuberia para organizar los flujos de machine learning. Este metodo promueva la eficiencia, repetibilidad y colaboracion en tus proyectos.
- Piensa en un pipeline(tuberia) comom una proceso de producion de una pelicula.
Un pipeline es un flujo de trabajo de gran nivel que organiza una serie de tareas o pasos requerido para crear un producto final.
- Ademas, ZenML representa una tuberia completa de flujo de trabajo de ML. Cada paso es una tuberia que envuelve la preparaciond e datos, ingeneria de datos, entrenamiento de modelo, evaluacion de modelo y finalmente despliegue de modelo.
- Cada paso de la tuberia depende de la salida del paso previo.
